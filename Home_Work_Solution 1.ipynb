```python
import pandas as pd
import random
```

```python
prices_file={
    'pain':1,
    'riz':2,
    'Viande_boeuf':5,
    'Viande_poulet':3,
    'Viande_dindon':4,
}
```

```python
class Customer():
    
    
    def __init__(self,registration_form,prices_file,id_customer=0):    
        self.registration_form=registration_form
        self.gender = None
        self.name=None
        self.birth_date=None
        self.address=None
        self.prices=prices_file
        self.id_achat=0
        self.promotion=0
        self.fidelity_state=False
        self.fidelity_gain=0
        
    def data_extracting(self):
        self.gender = 'M' if self.registration_form.split(' ')[0] == 'Mr' else 'F'
        self.name = self.registration_form.split(' ')[1]
        self.birthdate = self.registration_form.split('-')[1].split(' ')[2]
        self.address = self.registration_form.split('-')[-1]
    
    def data_featuring(self):
        
        if self.quantity>=50:
            self.client_type="Great"
            self.promotion=0.2
        else:
            self.client_type="Normal"
            
        
    #def set_product_price(self,product_name,new_price):
    #    self.product_name = new_price
           
    def buy_product(self,product_name,quantity):
        states=[False,True]
        self.product_name=product_name
        self.quantity=quantity
        self.product_price_unit=prices_file[product_name]
        self.fidelity_state=states[random.randint(0,1)]
        if self.fidelity_state and self.quantity>100:
            self.fidelity_gain+=0.01*self.product_price_unit*self.quantity
        

    def return_data_cleaned(self):
        
        return{
            'name': self.name,
            'gender': self.gender,
            'address':self.address,
            'quantity':self.quantity,
            'product_name':self.product_name,
            'product_price_unit':self.product_price_unit,
            'client_type':self.client_type,
            'promotion_granted':self.promotion,
            'fidelity_state':self.fidelity_state,
            'fidelity_gain':self.fidelity_gain
        }
            
```

```python
moussa=Customer('Mr Moussa - birthdate 1981/07/26 - 18 rue de pont neuf 06008',prices_file)
Jean=Customer('Mr Jean - birthdate 1900/07/22 - 18 rue de pont marcouville 06008',prices_file)
Bintou=Customer('Miss Bintou - birthdate 2000/07/26 - 34 Rés Osny 06008',prices_file)
Fanta=Customer('Miss Fanta - birthdate 1999/10/26 - 18 rue de Pontoise 06008',prices_file)
Lamine=Customer('Mr Lamine - birthdate 1950/07/26 - 45 rue de dalakan 06008',prices_file)
Ahmed=Customer('Mr Ahmed - birthdate 1995/07/13 - 90 rue de BKO 06008',prices_file)

customers=[
    moussa,
    Jean,
    Bintou,
    Fanta,
    Lamine,
    Ahmed
]
```

```python
customers[0].buy_product('pain',9)
customers[0].data_featuring()
customers[0].data_extracting()


customer_data=pd.DataFrame(columns=customers[0].return_data_cleaned().keys())

for i in range(random.randint(1000,3000)):

    client=customers[random.randint(0,len(customers)-1)]
    j=random.randint(0,100)
    client.buy_product(list(client.prices.keys())[j%5],random.randint(1,1000))
    client.data_featuring()
    client.data_extracting()
    df=pd.DataFrame([client.return_data_cleaned()], columns=client.return_data_cleaned())
    customer_data=pd.concat([customer_data,df])
```
